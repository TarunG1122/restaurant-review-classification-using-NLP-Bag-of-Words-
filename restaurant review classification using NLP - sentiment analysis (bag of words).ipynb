```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#remove doublr quotes 
#delimiter is use to detect spaces
df = pd.read_csv(r'C:\Users\tarun\Desktop\Restaurant_Reviews.tsv',delimiter='\t',quoting= 3)
df.dropna()
```

```python
#cleaning the text (removing punctuation)
#import regular ex
#nltk - remove all the words that dosent let words like (the , an ,)
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
#cleaned data will be added to corpus
corpus = []
#iterate for 1000 reviews
#sub is used to replace text - add square bracket too replace
# ^ - not in math
#stop words are words like {the , an , a}
for i in range(0,1000):
    review = re.sub('[^a-zA-Z]',' ',df['Review'][i])
    review = review.lower()
    review = review.split()
    # porterstemmer is used to stemming
    ps = PorterStemmer()
    #stopwords.words('english') - isto select english language
    #all_stopwords.remove('not') - is used because to exclude not word because can be used to  detect negative review
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = ''.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#creating bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
#sparx matrix
# fit_transform() will convert all the methods to columns
x = cv.fit_transform(corpus).toarray() # to array is used to convert elements to 0 and1
y = df.iloc[:-1]
np.reshape(y)
```

```python
#splitting train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```
